```
#Block E
import pandas as pd

# ======================================================================
# CHANGE THIS SECTION — Load YOUR dataset
# ======================================================================

# Describe what your dataset is about (used in the system prompt later)
DATASET_DESCRIPTION = "general knowledge documents"   # <-- CHANGE THIS: e.g. "COVID-19 research abstracts from 2020-2022"

# OPTION A: Load from HuggingFace datasets (recommended)
# Uncomment and adapt:
# from datasets import load_dataset
# ds = load_dataset("YOUR_DATASET_NAME_HERE", split="train")  # <-- CHANGE THIS
# documents = [str(row["text"]) for row in ds]  # adjust "text" to the correct column name

# OPTION B: Load from a CSV URL (HuggingFace raw file or Kaggle)
# Uncomment and adapt:
# DATASET_URL = "https://your-dataset-url-here.csv"  # <-- CHANGE THIS
# df = pd.read_csv(DATASET_URL)
# TEXT_COLUMN = "text"  # <-- CHANGE THIS: name of the column containing text
# documents = df[TEXT_COLUMN].dropna().astype(str).tolist()

# OPTION C: Paste your own list (for small datasets)
# Replace the strings below with real content from your domain:
documents = [
    """Strategic Feasibility of a Unified\nIntelligent Learning System for the CBSE\nEcosystem: Technical, Pedagogical, and\nRegulatory Perspectives\nThe current state of Indian secondary education, particularly within the Central Board of\nSecondary Education (CBSE) framework for Classes 6-12, is characterized by several structural constraints and pedagogical paradigms that warrant a comprehensive technological intervention.""",
    """The Technical Blueprint: Data Ingestion and the\nVectorization of CBSE Content\nThe proposed intelligent system's effectiveness is fundamentally predicated on the quality and organizational structure of its underlying knowledge base.""",
    """Retrieval-Augmented Generation (RAG) within\nCurricular Boundaries\nThe central innovation required for this system is a Curriculum-Bounded Retrieval-Augmented Generation (CB-RAG) architecture.""",
    """Engineering the Exam-Oriented Output Format\nA primary differentiator for the proposed system is its ability to generate responses in specific CBSE-style formats: Short Answers (2-3 marks), Long Answers (5-8 marks), and Numerical Problem Solutions.""",
    """Symbolic Rigor: Solving Numerical Problems in Math\nand Physics\nThe user requirement for numerical problems follows a specific logical flow: Given Formula Solution Final Answer.""",
    """Dual-Mode Pedagogy: Concept Explanation vs.\nRevision Mode\nThe intelligent system must cater to two distinct psychological states of a student: the deep learning phase and the rapid-revision phase.""",
    """Legal and Regulatory Landscape: Copyright and Fair\nDealing\nA significant hurdle in building a single intelligent system using NCERT, RD Sharma, and Lakhmir Singh is the legal and licensing framework."""
]

# ======================================================================
# Do NOT change below this line
# ======================================================================

assert len(documents) >= 5, f"You need at least 5 documents. You have {len(documents)}."
assert all(isinstance(d, str) and len(d.strip()) > 10 for d in documents), \
    "All documents must be non-empty strings with meaningful content."

print(f"Dataset loaded successfully!")
print(f"  Description   : {DATASET_DESCRIPTION}")
print(f"  Total documents: {len(documents)}")
print(f"  Avg doc length : {sum(len(d) for d in documents)//len(documents)} characters")
print(f"\nSample — first document preview:")
print(f"  {documents[0][:300]}...")
```

```
#Block B
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_core.documents import Document

# ======================================================================
# CHANGE THIS SECTION — Tune chunk parameters for your dataset
# ======================================================================

CHUNK_SIZE    = 500   # <-- CHANGE THIS: characters per chunk (tune to your doc length)
CHUNK_OVERLAP = 50    # <-- OPTIONALLY CHANGE: overlap between chunks (~10% of chunk size)
SOURCE_NAME   = "my_dataset"  # <-- CHANGE THIS: short label for your data source (e.g. "covid_papers", "legal_acts")

# ======================================================================
# Do NOT change below this line
# ======================================================================

splitter = RecursiveCharacterTextSplitter(
    chunk_size=CHUNK_SIZE,
    chunk_overlap=CHUNK_OVERLAP
)

langchain_docs = []
for doc_index, doc_text in enumerate(documents):
    chunks = splitter.split_text(doc_text)
    for chunk_index, chunk_text in enumerate(chunks):
        langchain_docs.append(
            Document(
                page_content=chunk_text,
                metadata={
                    "source": SOURCE_NAME,
                    "doc_index": doc_index,
                    "chunk_index": chunk_index
                }
            )
        )

print(f"Chunking complete!")
print(f"  Original documents  : {len(documents)}")
print(f"  Total chunks created: {len(langchain_docs)}")
print(f"  Avg chunks per doc  : {len(langchain_docs) / len(documents):.1f}")
print(f"  Chunk size          : {CHUNK_SIZE} chars | Overlap: {CHUNK_OVERLAP} chars")
print(f"\nSample — Chunk 0 preview:")
print(f"  Content : {langchain_docs[0].page_content[:200]}...")
print(f"  Metadata: {langchain_docs[0].metadata}")
```

```
#Block G
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

# ======================================================================
# OPTIONALLY CHANGE — Embedding model (default works for most datasets)
# ======================================================================

EMBEDDING_MODEL = "sentence-transformers/all-MiniLM-L6-v2"  # <-- OPTIONALLY CHANGE: see table above

# ======================================================================
# Do NOT change below this line
# ======================================================================

print(f"Loading embedding model: {EMBEDDING_MODEL}")
print("(First run downloads ~90 MB — subsequent runs use cache)")

embeddings = HuggingFaceEmbeddings(
    model_name=EMBEDDING_MODEL,
    encode_kwargs={"normalize_embeddings": True}
)

print(f"Building FAISS vector store from {len(langchain_docs)} chunks...")
print("(This may take a minute for large datasets)")

vectorstore = FAISS.from_documents(langchain_docs, embeddings)

# Quick sanity check
test_results = vectorstore.similarity_search("test query", k=1)
print(f"\nFAISS index built successfully!")
print(f"  Embedding model    : {EMBEDDING_MODEL}")
print(f"  Embedding dimensions: 384")
print(f"  Chunks indexed     : {len(langchain_docs)}")
print(f"  Test retrieval     : returned {len(test_results)} result(s) ✓")
print(f"\nSample retrieved chunk:")
print(f"  {test_results[0].page_content[:200]}...")
```

```
#Block D
from langchain_community.llms import Ollama

# ======================================================================
# OPTIONALLY CHANGE — Choose the Ollama model and parameters
# ======================================================================

OLLAMA_MODEL = "llama3.2:1b"  # <-- OPTIONALLY CHANGE: model name (see table above)
TEMPERATURE  = 0.1            # <-- OPTIONALLY CHANGE: 0=deterministic, 1=creative
MAX_TOKENS   = 400            # <-- OPTIONALLY CHANGE: max response length in tokens

# ======================================================================
# Do NOT change below this line
# ======================================================================

print(f"Pulling Ollama model: {OLLAMA_MODEL}")
print("(This may take 1–3 minutes on first run — the model is being downloaded)")
import subprocess
subprocess.run(["ollama", "pull", OLLAMA_MODEL], check=True)
print(f"Model downloaded!")

llm = Ollama(
    model=OLLAMA_MODEL,
    temperature=TEMPERATURE,
    num_predict=MAX_TOKENS,
)

print(f"\nOllama LLM configured:")
print(f"  Model       : {OLLAMA_MODEL}")
print(f"  Temperature : {TEMPERATURE}")
print(f"  Max tokens  : {MAX_TOKENS}")
print("\nTesting LLM connection — sending a test prompt...")
test_resp = llm.invoke("Reply with exactly three words: I am ready")
print(f"LLM test response: {test_resp.strip()}")
print("Ollama LLM is ready!")
```

```
#Block H
from langchain_core.prompts import PromptTemplate

# ======================================================================
# OPTIONALLY CHANGE — Customise the tone or format of answers
# ======================================================================

# Add any extra instructions for how the LLM should respond
CUSTOM_INSTRUCTION = "Be concise and precise. Quote directly from the documents when possible."  # <-- OPTIONALLY CHANGE

# ======================================================================
# Do NOT change below this line
# ======================================================================

PROMPT_TEMPLATE = f"""You are an expert assistant specialised in {DATASET_DESCRIPTION}.
Your task is to answer questions based ONLY on the context documents provided below.
Do NOT use your training knowledge. If the answer is not in the context, say: "I cannot find this information in the provided documents."
{CUSTOM_INSTRUCTION}

Context documents:
{{context}}

Question: {{question}}

Answer (grounded in the context above):"""

prompt_template = PromptTemplate(
    input_variables=["context", "question"],
    template=PROMPT_TEMPLATE
)

print("Prompt template created!")
print("\nTemplate preview:")
print("-" * 50)
sample = PROMPT_TEMPLATE.replace("{context}", "[RETRIEVED CHUNKS GO HERE]").replace("{question}", "[USER QUESTION GOES HERE]")
print(sample)
print("-" * 50)
print("\nVariables that will be filled at query time:")
print("  {context}  ← top-K retrieved document chunks")
print("  {question} ← user's query")
```

```
#Block A
from langchain_classic.chains import RetrievalQA
# ======================================================================
# OPTIONALLY CHANGE — number of chunks to retrieve per question
# ======================================================================

TOP_K = 3   # <-- OPTIONALLY CHANGE: how many document chunks to retrieve (try 3, 4, or 5)

# ======================================================================
# Do NOT change below this line
# ======================================================================

retriever = vectorstore.as_retriever(
    search_type="similarity",
    search_kwargs={"k": TOP_K}
)

rag_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    return_source_documents=True,
    chain_type_kwargs={"prompt": prompt_template}
)

print(f"RAG chain assembled successfully!")
print(f"Retriever: FAISS similarity search, k={TOP_K}")
print(f"LLM: Ollama local model")
print(f"Chain type: stuff (all chunks stuffed into one prompt)")
print("The pipeline is ready for queries!")
```

```
#Block F
# ======================================================================
# Verify the pipeline is working correctly with test queries
# ======================================================================

test_queries = [
    "What are some key features mentioned in the document?",
    "Tell me about the system architecture.",
    "Describe the technical requirements."
]

print("Running test queries...")
for i, query in enumerate(test_queries, 1):
    result = rag_chain.invoke({"query": query})
    print(f"Test query {i} successful ✓")

print(f"\nSUCCESS: RAG Pipeline is fully operational!")
print(f"You can now run custom queries using rag_chain.invoke({{\"query\": \"YOUR QUESTION HERE\"}})")
```
