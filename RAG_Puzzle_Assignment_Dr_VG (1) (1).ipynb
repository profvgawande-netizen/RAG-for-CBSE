```python
#Block B
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_core.documents import Document

# ======================================================================
# CHANGE THIS SECTION — Tune chunk parameters for your dataset
# ======================================================================

CHUNK_SIZE    = 500   # <-- CHANGE THIS: characters per chunk (tune to your doc length)
CHUNK_OVERLAP = 50    # <-- OPTIONALLY CHANGE: overlap between chunks (~10% of chunk size)
SOURCE_NAME   = "my_dataset"  # <-- CHANGE THIS: short label for your data source (e.g. "covid_papers", "legal_acts")

# ======================================================================
# Do NOT change below this line
# ======================================================================

splitter = RecursiveCharacterTextSplitter(
    chunk_size=CHUNK_SIZE,
    chunk_overlap=CHUNK_OVERLAP
)

langchain_docs = []
for doc_index, doc_text in enumerate(documents):
    chunks = splitter.split_text(doc_text)
    for chunk_index, chunk_text in enumerate(chunks):
        langchain_docs.append(
            Document(
                page_content=chunk_text,
                metadata={
                    "source": SOURCE_NAME,
                    "doc_index": doc_index,
                    "chunk_index": chunk_index
                }
            )
        )

print(f"Chunking complete!")
print(f"  Original documents  : {len(documents)}")
print(f"  Total chunks created: {len(langchain_docs)}")
print(f"  Avg chunks per doc  : {len(langchain_docs) / len(documents):.1f}")
print(f"  Chunk size          : {CHUNK_SIZE} chars | Overlap: {CHUNK_OVERLAP} chars")
print(f"\nSample — Chunk 0 preview:")
print(f"  Content : {langchain_docs[0].page_content[:200]}...")
print(f"  Metadata: {langchain_docs[0].metadata}")
```

```python
#Block G
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

# ======================================================================
# OPTIONALLY CHANGE — Embedding model (default works for most datasets)
# ======================================================================

EMBEDDING_MODEL = "sentence-transformers/all-MiniLM-L6-v2"  # <-- OPTIONALLY CHANGE: see table above

# ======================================================================
# Do NOT change below this line
# ======================================================================

print(f"Loading embedding model: {EMBEDDING_MODEL}")
print("(First run downloads ~90 MB — subsequent runs use cache)")

embeddings = HuggingFaceEmbeddings(
    model_name=EMBEDDING_MODEL,
    encode_kwargs={"normalize_embeddings": True}
)

print(f"Building FAISS vector store from {len(langchain_docs)} chunks...")
print("(This may take a minute for large datasets)")

vectorstore = FAISS.from_documents(langchain_docs, embeddings)

# Quick sanity check
test_results = vectorstore.similarity_search("test query", k=1)
print(f"\nFAISS index built successfully!")
print(f"  Embedding model    : {EMBEDDING_MODEL}")
print(f"  Embedding dimensions: 384")
print(f"  Chunks indexed     : {len(langchain_docs)}")
print(f"  Test retrieval     : returned {len(test_results)} result(s) ✓")
print(f"\nSample retrieved chunk:")
print(f"  {test_results[0].page_content[:200]}...")
```

```python
#Block D
from langchain_community.llms import Ollama

# ======================================================================
# OPTIONALLY CHANGE — Choose the Ollama model and parameters
# ======================================================================

OLLAMA_MODEL = "llama3.2:1b"  # <-- OPTIONALLY CHANGE: model name (see table above)
TEMPERATURE  = 0.1            # <-- OPTIONALLY CHANGE: 0=deterministic, 1=creative
MAX_TOKENS   = 400            # <-- OPTIONALLY CHANGE: max response length in tokens

# ======================================================================
# Do NOT change below this line
# ======================================================================

print(f"Pulling Ollama model: {OLLAMA_MODEL}")
print("(This may take 1–3 minutes on first run — the model is being downloaded)")
import subprocess
subprocess.run(["ollama", "pull", OLLAMA_MODEL], check=True)
print(f"Model downloaded!")

llm = Ollama(
    model=OLLAMA_MODEL,
    temperature=TEMPERATURE,
    num_predict=MAX_TOKENS,
)

print(f"\nOllama LLM configured:")
print(f"  Model       : {OLLAMA_MODEL}")
print(f"  Temperature : {TEMPERATURE}")
print(f"  Max tokens  : {MAX_TOKENS}")
print("\nTesting LLM connection — sending a test prompt...")
test_resp = llm.invoke("Reply with exactly three words: I am ready")
print(f"LLM test response: {test_resp.strip()}")
print("Ollama LLM is ready!")
```

```python
#Block H
from langchain_core.prompts import PromptTemplate

# ======================================================================
# OPTIONALLY CHANGE — Customise the tone or format of answers
# ======================================================================

# Add any extra instructions for how the LLM should respond
CUSTOM_INSTRUCTION = "Be concise and precise. Quote directly from the documents when possible."  # <-- OPTIONALLY CHANGE

# ======================================================================
# Do NOT change below this line
# ======================================================================

PROMPT_TEMPLATE = f"""You are an expert assistant specialised in {DATASET_DESCRIPTION}.
Your task is to answer questions based ONLY on the context documents provided below.
Do NOT use your training knowledge. If the answer is not in the context, say: "I cannot find this information in the provided documents."
{CUSTOM_INSTRUCTION}

Context documents:
{{context}}

Question: {{question}}

Answer (grounded in the context above):"""

prompt_template = PromptTemplate(
    input_variables=["context", "question"],
    template=PROMPT_TEMPLATE
)

print("Prompt template created!")
print("\nTemplate preview:")
print("-" * 50)
sample = PROMPT_TEMPLATE.replace("{context}", "[RETRIEVED CHUNKS GO HERE]").replace("{question}", "[USER QUESTION GOES HERE]")
print(sample)
print("-" * 50)
print("\nVariables that will be filled at query time:")
print("  {context}  ← top-K retrieved document chunks")
print("  {question} ← user's query")
```

```python
#Block A
from langchain_classic.chains import RetrievalQA
# ======================================================================
# OPTIONALLY CHANGE — number of chunks to retrieve per question
# ======================================================================

TOP_K = 3   # <-- OPTIONALLY CHANGE: how many document chunks to retrieve (try 3, 4, or 5)

# ======================================================================
# Do NOT change below this line
# ======================================================================

retriever = vectorstore.as_retriever(
    search_type="similarity",
    search_kwargs={"k": TOP_K}
)

rag_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    return_source_documents=True,
    chain_type_kwargs={"prompt": prompt_template}
)

print(f"RAG chain assembled successfully!")
print(f"Retriever: FAISS similarity search, k={TOP_K}")
print(f"LLM: Ollama local model")
print(f"Chain type: stuff (all chunks stuffed into one prompt)")
print("The pipeline is ready for queries!")
```

```python
#Block F
# ======================================================================
# CHANGE THIS SECTION — Write your own question about YOUR dataset
# ======================================================================

MY_QUESTION = "What is the main topic discussed in these documents?"  # <-- CHANGE THIS: ask something answerable from your data

# ======================================================================
# Do NOT change below this line
# ======================================================================

print(f"Sending query to RAG system...")
print(f"Question: {MY_QUESTION}")
print("=" * 60)

result = rag_chain.invoke({"query": MY_QUESTION})

print(f"\nANSWER:")
print(result["result"].strip())

print(f"\nSOURCE DOCUMENTS RETRIEVED ({len(result['source_documents'])} chunks):")
for i, doc in enumerate(result["source_documents"], 1):
    print(f"\n  [{i}] Source: {doc.metadata}")
    print(f"       Preview: {doc.page_content[:150]}...")

print("\n" + "=" * 60)
print("If the answer looks wrong or vague, check:")
print("  1. Are your documents actually loaded? (Step 2)")
print("  2. Is your chunk size appropriate? (Step 3)")
print("  3. Try increasing TOP_K in the chain step")
print("  4. Is this question answerable from the documents?")
```
